```python
import nibabel as nib
import numpy as np
import os
```

```python
ruta = '/home/jorgetbm/atlas/controlTemplateSqueeze.nii'
atlas_cinco = nib.load(ruta).get_fdata()
atlas_cinco = atlas_cinco[..., np.newaxis]
atlas_cinco = atlas_cinco[np.newaxis, ...]
```

```python
!wget https://surfer.nmr.mgh.harvard.edu/ftp/data/voxelmorph/models/vxm_dense_brain_T1_3D_mse.h5
```

```python
inshape = atlas_cinco.shape[1:-1]
nb_feats = atlas_cinco.shape[-1]

print(inshape)
print(nb_feats)

gpu = 0
device, nb_devices = vxm.tf.utils.setup_device(gpu)

ruta_carpeta = '/home/jorgetbm/imagenesCNresized/'

lista_imagenes = os.listdir(ruta_carpeta)

lista_moved = []
lista_warp = []

with tf.device(device):
    # load model and predict
    config = dict(inshape=inshape, input_model=None)

    for nombre_imagen in lista_imagenes:
      ruta = ruta_carpeta + nombre_imagen
      print(ruta)
      imagen = nib.load(ruta).get_fdata()
      imagen = imagen[..., np.newaxis]
      imagen = imagen[np.newaxis, ...]
      imagen = (imagen - np.min(imagen)) / (np.max(imagen) - np.min(imagen))
      imagen.shape
      warp = vxm.networks.VxmDense.load('/home/jorgetbm/vxm_dense_brain_T1_3D_mse.h5', **config).register(imagen, atlas_cinco)
      moved = vxm.networks.Transform(inshape, nb_feats=nb_feats).predict([imagen, warp])
      lista_moved.append(moved)
      lista_warp.append(warp)
```

```python
print(lista_warp[0].shape)
print(len(lista_warp))
```

```python
from scipy.ndimage import zoom

new_shape = (1, 176, 208, 176, 3)
new_images = []

gpu = 0
device, nb_devices = vxm.tf.utils.setup_device(gpu)

with tf.device(device):
  for imagen_warp in lista_warp:
    new_image = zoom(imagen_warp, (new_shape[0]/imagen_warp.shape[0], new_shape[1]/imagen_warp.shape[1], new_shape[2]/imagen_warp.shape[2], new_shape[3]/imagen_warp.shape[3], 1))
    print(new_image.shape)
    new_images.append(new_image)
```
