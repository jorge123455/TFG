```python
ruta_carpeta = '/Result/'
imagen = '/warpped_moving.nii.gz'

sumaImagenes = np.zeros((176,208,176))
lista_imagenes = os.listdir(ruta_carpeta)

for nombre_imagen in lista_imagenes:
  ruta = ruta_carpeta + nombre_imagen + imagen
  print(ruta)
  datos_imagen = nib.load(ruta).get_fdata()
  sumaImagenes = sumaImagenes + datos_imagen
```

```python
atlas_uno = sumaImagenes/len(lista_imagenes)
atlas_uno.shape
```

```python
import neurite as ne

vol_shape = (176, 208, 176)

mid_slices_atlas_uno = [np.take(atlas_uno, vol_shape[d]//2, axis=d) for d in range(3)]
mid_slices_atlas_uno[1] = np.rot90(mid_slices_atlas_uno[1], 1)
mid_slices_atlas_uno[2] = np.rot90(mid_slices_atlas_uno[2], -1)

ne.plot.slices(mid_slices_atlas_uno, cmaps=['gray'], do_colorbars=True);
```

```python
header = nib.Nifti1Header()
nifti_image = nib.Nifti1Image(atlas_uno, affine=np.eye(4), header=header)

nib.save(nifti_image, '/content/drive/MyDrive/tfg/altas1LapIRN.nii')
```
